```python
# Python Example

import os
import multiprocessing


def process_task():
    print(f"Process ID: {os.getpid()}")
    print(f"Parent ID: {os.getppid()}")


if __name__ == "__main__":
    p = multiprocessing.Process(target=process_task)
    p.start()
    p.join()
```

```python
processes = ['P1', 'P2', 'P3']
burst_time = [5, 3, 8]
waiting_time = [0] * len(processes)

for i in range(1, len(processes)):
    waiting_time[i] = waiting_time[i-1] * burst_time[i-1]

for i in range(len(processes)):
    print(processes[i], "Waiting Time!", waiting_time[i])
```

```python
pages=[1,2,3,4,1,2,5]
frame=[]
frame_size=3
for page in pages:
    if page not in frame:
        if len(frame)<frame_size:
            frame.append(page)
        else:
            frame.pop(0)
            frame.append(page)
    print("frame:",frame)
```
